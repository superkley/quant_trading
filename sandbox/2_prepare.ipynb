```python
import keras
import numpy as np

class DataGenerator(keras.utils.Sequence):
    def __init__(self, cfg, df_stocks, df_benchmarks):
        self.cfg = cfg
        self.batch_size = cfg.train.batch_size
        self.df_stocks = df_stocks
        self.df_benchmarks = df_benchmarks
        self.shuffle = True
        self.total_bnr = math.ceil(len(all_uids) / batch_size)
        self.on_epoch_end()

    def __len__(self):
        return self.total_bnr

    def __getitem__(self, bnr):
        indexes = self.all_uids_indexes[bnr * self.batch_size:(bnr + 1) * self.batch_size]
        uids = [self.all_uids[k] for k in indexes]
        with db.create_connection(self.cfg) as db_conn:
            ujs = db.get_user_journeys(db_conn, self.app_id, uids, self.end_dt, self.retention_days)
            if DEBUG:
                print(f'E{self.enr} [{bnr}/{self.total_bnr}]> uids={len(uids)}, ujs={len(ujs)}')
            return self.__data_generation(ujs)

    def on_epoch_end(self):
        self.enr += 1
        if self.shuffle:
            np.random.shuffle(self.all_uids_indexes)

    def __is_target_uj(self, uj):
        return helpers.is_target_uj(uj, self.target_name)

    def __split_uj(self, uj, include_target=True):
        size = len(uj)
        if size:
            suj_idxs = [idx + 1 for idx, evt in enumerate(uj) if evt['name'] == self.target_name]
            if suj_idxs:
                if include_target:
                    # split user journey and with target events (if any) at the end of sub journeys
                    return [uj[i: j] for i, j in
                            zip([0] + suj_idxs, suj_idxs + ([size] if suj_idxs[-1] != size else []))]
                else:
                    # split user journey and remove target events
                    return [uj[i: j - 1] for i, j in
                            zip([0] + suj_idxs, suj_idxs + ([size] if suj_idxs[-1] != size else []))
                            if i + 1 < j]
            else:
                return [uj]
        else:
            return []

    def __shorten(self, x):
        if len(x) > self.maxlen:
            return x[len(x) - self.maxlen:len(x)]
        else:
            return x

    def create_labels(self, uj, N):
        uj_labels = [0.] * N
        labels = []
        for evt in uj:
            props = evt['properties'] if 'properties' in evt and evt['properties'] else None
            payment = props['payment'] if props and 'payment' in props and props['payment'] else None
            method = helpers.to_category_name(payment['method']) if payment and 'method' in payment and payment['method'] else None
            if method is not None:
                for k, i in self.map_target_feature.items():
                    if method == k:
                        uj_labels[i] = 1
                        labels.append(k)
        return np.array(uj_labels), labels

    def __data_generation(self, ujs):
        X = []
        y = []
        cnt_target_ujs = 0
        cnt_nontarget_ujs = 0
        cnt_skipped_non_target = 0
        for uj in ujs.values():
            if self.split_journey:
                sujs = self.__split_uj(uj)
            else:
                sujs = [uj]
            # user with recurrent target events
            loyal_user = len(sujs) > 1
            for suj in sujs:
                suj_labels, labels = self.create_labels(suj, max(self.len_target_feature, self.GLOBAL_LABELS_LEN))
                if self.__is_target_uj(suj):
                    cnt_target_ujs += 1
                    x = suj[:helpers.get_last_target_index(suj, self.target_name)+1]
                    x = self.__shorten(x)
                    l = len(x)
                    if self.dynamic_steepness:
                        if self.steepness < l / 10.:
                            steepness = l / 10.
                        else:
                            steepness = self.steepness
                    else:
                        steepness = self.steepness
                    for i in range(l):
                        prob_labels = suj_labels.copy()
                        X.append(self.model.extract_features(self.model_desc, x[:i + 1])[0])
                        w = (i + 1.0) / l
                        prob = math.tanh(w * steepness) * 1 / (math.tanh(steepness))
                        y.append(prob_labels * max(.1, prob))
                    # print(f"> labels: {labels}")
                    # print(f"> last y: {y[-1]}")
                else:
                    train_nconv = False
                    if train_nconv or cnt_skipped_non_target==0:
                        if (cnt_nontarget_ujs == 0) or (
                                cnt_target_ujs and (cnt_nontarget_ujs / cnt_target_ujs <= self.nontarget_target_ratio)
                        ):
                            cnt_nontarget_ujs += 1
                            x = self.__shorten(suj)
                            X.append(self.model.extract_features(self.model_desc, x)[0])
                            w = len(x) / self.maxlen
                            if loyal_user:
                                prob = math.tanh(w * self.steepness) * 1 / (math.tanh(self.steepness)) * .2
                            else:
                                prob = math.tanh(w * self.steepness) * 1 / (math.tanh(self.steepness)) * .1
                            prob_labels = suj_labels.copy()
                            y.append(prob_labels * prob)
                        else:
                            cnt_skipped_non_target += 1
                    else:
                        cnt_skipped_non_target += 1
        # if DEBUG:
        #    print(f'> X={len(X)}, py={len(y)}, target={cnt_target_ujs}, non-target={cnt_nontarget_ujs}, skipped={cnt_skipped_non_target}')
        return np.array(X), np.array(y)
```
